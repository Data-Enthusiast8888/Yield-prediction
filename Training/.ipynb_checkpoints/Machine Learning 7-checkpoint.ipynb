### Train, Validation and Test datasets

```python
#Divide the dataset into dependent and independent variables 
X = soil_data[soil_data.columns[:-1]].values
y = soil_data[soil_data.columns[-1]].values
```

```python
# split the dataset

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state =1111)
X_train, X_valid, y_train, y_valid = train_test_split(X_train,y_train, test_size=0.2, random_state=1111)

#check the size of the splits 
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Validation  set size: {X_valid.shape[0]} samples")
print(f"Test set Size: {X_test.shape[0]} samples")
```

```python
print(sum(y_train==0))
print(sum(y_train==1))
print(sum(y_train==2))
```

```python
def R_sampling(oversample=False):
    if oversample:
        ros = RandomOverSampler()
        X,y = ros.fit_resample(X,y)
        return X,y
```

```python
R_sampling(oversample=False)
```

```python
# Scale the dataset
sc =StandardScaler()

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

## SVM

```python
from sklearn import svm

from sklearn.svm import SVC

from sklearn.metrics import classification_report
```

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
svm_model.score(X,y)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

knn_model = KNeighborsClassifier(n_neighbors=3)

knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
knn_model.score(X,y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeRegressor
import warnings
warnings.filterwarnings("ignore")

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'logistic_regression' : {
            'model': LogisticRegression(),
            "params":{},
        },
        'KNeighborsClassifier': {
            'model': KNeighborsClassifier(n_neighbors=3),
             "params":{},
            
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            "params":{},
        },

        'support_vector_machine':{

             "model":svm.SVC(),
             "params":{},
        },

      'Naive_bayes':{

             "model":GaussianNB(),
             "params":{},
        },

    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'],cv=cv, return_train_score=True)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score', 'best_param'])

find_best_model_using_gridsearchcv(X,y)
```
